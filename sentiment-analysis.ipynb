```python
# removing the Html tags
import re

def remove_html_tags(text):
  clean = re.compile('<.*?>')
  return re.sub(clean, '', text)

remove_html_tags(df['review'].iloc[0])

# The code seems to work properly
```

```python
df.info()
```

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
# Conveting the text to lowarcase
def Lower(text:str)->str:
    return text.lower()

Lower(df['review'].iloc[0])
# The code seems to work properly
```

```python
# Now applying it to the all rows
df['review'] = df['review'].apply(Lower)
```

```python
# removing the alphanumeric characters
def remove_special(text):
    x = ''
    for i in text:
        if i.isalnum():
            x = x+i
        else:
            x = x + ' '
    return x
```

```python
# Now applying this to the whole column
df['review'] = df['review'].apply(remove_special)
```

```python
# Removal of the stopwords
import nltk
from nltk.corpus import stopwords

# Download the stopwords list
nltk.download('stopwords')

def remove_stopwords(text:str)->str:
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in text.split() if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
# Apply the above function to the column
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df['review'].iloc[0]
# The above code seems to work properly
```

```python
# now applying the stemming
from nltk.stem import PorterStemmer

# Download NLTK data (if not already downloaded)
nltk.download('punkt')

def stemm(text:str)->str:
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in text.split()]
    return ' '.join(stemmed_words)
```

```python
# Appling the stemm function to the review column
df['review'] = df['review'].apply(stemm)
```

```python
df['review'].iloc[0] # The code seems to be working properly
```

```python
# Converting sentiment column to 0's and 1's
# 1 for positive review and 0 for the negative review
def convert(word):
    if(word=='positive'):
        return 1
    return 0;
convert(df['sentiment'].iloc[0])
# The above code seems to be working properly
```

```python
df['sentiment'] = df['sentiment'].apply(convert)
```

```python
df
# Now our data is much more preprocessed
# Now we can move forward by applying bag of words technique
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectoriser = CountVectorizer()
```

```python
X = vectoriser.fit_transform(df['review']).toarray() # This is our feature
print(X[0:5])
print("Shape of X is : ", X.shape)
```

```python
y = df['sentiment'].values
print(y[0:5])
print("The shape of y is : ",y.shape)
```

```python
# Creating training and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Importing models from the sklearn library
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()

# Fitting the models
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
clf3.fit(X_train, y_train)

print('Model Training compleat')
```

```python
# now let's make some predictions
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
y_pred3 = clf3.predict(X_test)
```

```python
# Now lets calculate the individual accuracies
from sklearn.metrics import accuracy_score
print("Accuracy Score for GaussianNB : ", accuracy_score(y_test, y_pred1))
print("Accuracy Score for MultinomialNB : ", accuracy_score(y_test, y_pred2))
print("Accuracy Score for BernoulliNB : ", accuracy_score(y_test, y_pred3))
```
